## Problem with K-means

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles,make_moons
```

```python
X,Y = make_moons(n_samples=200,noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
km.fit(X)
```

```python
centers = km.cluster_centers_
label = km.labels_

plt.scatter(X[:,0],X[:,1],c=label)
plt.scatter(centers[:,0],centers[:,1],marker="*",color="red")
plt.show()
```

### k-means does  fail on data set which are quite complex or which are not linearly seperable. It will not work on such cases.

### For these kind of patterns we are going to use DBSCAN

# DBSCAN
- **Density Based Spatial Clustering of Applications with Noise**
- Based upon on the idea, that a cluster is a high density area surrounded by low density region
- Starts by exploring a small area, if density is "good enough" it considered as part of clustered as part of cluster, and explores neigbours and so on, so it groups regions based upon density!

```python
plt.scatter(X[:,0],X[:,1],c=label)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2,min_samples=6)
```

```python
Y_pred = dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y_pred)
plt.show()
```
